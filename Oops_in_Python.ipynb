#Constructor:
* Constructor is a method
* a constructor is a special type of method that is automatically invoked when an object of a class is created. The primary purpose of a constructor is to initialize the newly created object.
* Constructor is called automatically just after creating an object for a class

**The __init__
* method in Python is a special method that is used to initialize a newly created object. It's often referred to as the constructor because it's called automatically when a new instance of a class is created. The __init__ method allows you to set the initial state of the object by assigning values to its attributes.

Key Features of __init__:
Automatic Invocation:

The __init__ method is automatically called when a new object of a class is instantiated.
Initialization:

The main purpose of __init__ is to initialize the object's attributes with the values provided during object creation.
Self Parameter:

The first parameter of __init__ is always self, which is a reference to the instance of the class being created. It allows you to access and modify the object's attributes.
No Return Value:

The __init__ method doesn't return any value (i.e., it must return None). Its role is to prepare the new object, not to return a value.
Optional Parameters:

__init__ can take additional parameters, allowing you to pass arguments to initialize the object's attributes.

# What is Inheritance in Object-Oriented Programming (OOP)?
* Inheritance is a fundamental concept in object-oriented programming (OOP) that allows one class (called a child or subclass) to inherit the properties and behaviors (attributes and methods) of another class (called a parent or superclass).

* Inheritance is used to create a new class based on an existing class, which promotes code reuse and helps organize code more efficiently.

# Key Concepts of Inheritance:
* Parent (Super) Class: The class whose properties and methods are inherited by another class.
* Child (Sub) Class: The class that inherits the properties and methods from the parent class.
* Method Overriding: The child class can provide a specific implementation of a method that is already defined in its parent class.

# Why is Inheritance Useful?
* Code Reuse: Instead of writing the same code multiple times, you can write it once in a parent class and reuse it in child classes.
* Logical Organization: Inheritance allows you to model real-world relationships, making your code easier to understand and maintain.
* Extensibility: You can extend the functionality of an existing class by adding new methods or overriding existing ones in the child class.

```python
#Example of Inheritance
#Let's take a simple example of a parent class called Animal
#and a child class called Dog.

# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} makes a sound"

# Child class
class Dog(Animal):         #class dog(animal)
    def speak(self):
        return f"{self.name} barks"

class cat(Animal):
    def speak(self):
        return f"{self.name} Meow"

# Creating an object of the parent class
animal = Animal("Generic Animal")
print(animal.speak())
# Creating an object of the child class
dog = Dog("Max")
print(dog.speak())
Cat=cat("Jasmine")
print(Cat.speak())
```

# What is Polymorphism in OOP?

* It also allows methods to perform different functions based on the object that is calling them. In simple words, polymorphism means "many forms," and it enables a single interface to represent different underlying forms (data types).
* Polymorphism is achieved in two main ways:

* Method Overriding (Runtime Polymorphism): This occurs when a subclass provides a specific implementation of a method that is already defined in its superclass.
* Method Overloading (Compile-time Polymorphism): This occurs when multiple methods in the same class have the same name but different parameters.

- Example 1: Method Overriding (Runtime Polymorphism)
Let's consider an example where we have a base class called Animal and two subclasses called Dog and Cat. Both subclasses override the sound() method.

```python
# Method overriding
class Animal:
    def sound(self):
        return "Some Generic Sound"

class Dog:
    def sound(self):
        return "Barks"
    
class Cat:
    def sound(self):
        return "Meow"

#Creating objects
animal = Animal()
cat = Cat()
dog = Dog()

#call methods
print(animal.sound())
print(cat.sound())
print(dog.sound())
```

```python
class Worker:
    def work(self):
       print("Working")
    def perform_task(self):
       print("Performing Task: ", end='')
       self.work()

class DeliveryMan(Worker):
    def work(self):
       print("Delivering Goods")

class LumberJack(Worker):
     def work(self):
        print("Cutting Woods")

deliveryMan = DeliveryMan()
lumberJack = LumberJack()

deliveryMan.perform_task()
lumberJack.perform_task()
```

Example 2: Method Overloading (Compile-time Polymorphism)

# Encapsulation in Object-Oriented Programming (OOP)

- Encapsulation is one of the fundamental concepts in OOP. It refers to the bundling of data (variables) and methods (functions) that operate on the data into a single unit or class. It also involves restricting access to some of the object's components, meaning that some data and methods in a class can be hidden from outside access. This is typically done using access modifiers like private, protected, and public.

- Key Benefits of Encapsulation:

1. Control: Encapsulation allows you to control the access to your data. You can restrict access to certain properties and methods of a class, ensuring that they are not altered or accessed inappropriately.
2. Modularity: It helps in dividing the code into smaller, manageable pieces (modules).
3. Security: By hiding sensitive data, you protect the integrity of the object.
4. Ease of Maintenance: Encapsulation makes the code more flexible and easier to maintain.
