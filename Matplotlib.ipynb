# Matplotlib in Python

   - Matplotlib is a powerful and versatile open-source plotting library for Python, designed to help users visualize data in a variety of formats. Developed by John D. Hunter in 2003, it enables users to graphically represent data, facilitating easier analysis and understanding. If you want to convert your boring data into interactive plots and graphs, Matplotlib is the tool for you.

   - Key Features
        - **Versatile plotting:**
            Matplotlib supports a wide range of plot types, including line plots, scatter plots, bar charts, histograms, pie charts, 3D plots, and more.
        - **Customization:**
            It allows for extensive customization of plot elements, such as colors, labels, gridlines, and styles.
        - **Integration:**
            Matplotlib works well with other Python libraries like NumPy and Pandas, making it easy to visualize data from various sources.
        - **Output formats:**
            It can export plots in various formats, including PNG, JPG, PDF, and SVG.
        - **Interfaces:**
            Matplotlib offers both a MATLAB-like interface through pyplot and an object-oriented interface for more advanced customization.

# Matplotlib subplot
Subplots are smaller axes (plots) arranged within a single figure allowing users to compare data side by side or stack visualizations for better insights. plt.subplots() method provides an easy way to define the layout of subplots.
- The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0,10,1)  #[0,1,2,3,4,....,9]
y1 = 2*x+5             #[5,7,..........,23]
y2 = 3*x+10            #[10,13,.........,37]

#Create two or more plots in a fig
plt.subplot(1,2,1)
#PLotting the points
plt.plot(x,y1,color ="r")
plt.title('Graph1')

plt.subplot(1,2,2)
plt.plot(x,y2,color='b')
plt.title('Graph2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0,10,1)  #[0,1,2,3,4,....,9]
y1 = 2*x+5             #[5,7,..........,23]
y2 = 3*x+10            #[10,13,.........,37]

#Create two or more plots in a fig
plt.subplot(2,1,1)
#PLotting the points
plt.plot(x,y1,linewidth = 2.0,linestyle = "-.", color = 'r',alpha = 0.8,marker ='*')

plt.grid(True)
plt.title('Graph1')

plt.subplot(2,1,2)
plt.plot(x,y2,linewidth = 2.0,linestyle = "-.", color = 'm',alpha = 0.8,marker ='s')
plt.title('Graph2')
plt.grid(True)
plt.show()
```

# Bar Chart OR Bar Plot
A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. It can be created using the bar() method.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#creating the dataset
data = {'apples':20,'Mangoes':15,'Lemon':30,'Oranges':10}  #Keys= [apples, Mangoes, Lemon,Oranges]
names = list(data.keys())         #['apples','mangoes','Lemon','Oranges']                                 #VAlues=[20,15,30,10]
values = list(data.values())      #[20,15,30,10]

#fig = plt.figure(figsize = (5,10))
#Creating the bar plot
plt.bar(names,values,color=['gold','c','b','k'])
plt.show()
```

# Histogram Plot
Histogram is basically used to represent data in the form of some groups. It is a type of bar plot where the X-axis represents the bin ranges while the Y-axis gives information about frequency. To create a histogram the first step is to create a bin of the ranges, then distribute the whole range of the values into a series of intervals and count the values which fall into each of the intervals. Bins are clearly identified as consecutive, non-overlapping intervals of variables.

- a histogram chart basically used to visualize the frequency distribution of numerical data by grouping it into bins or intervals.

```python
import matplotlib.pyplot as plt
%matplotlib inline
#prepraring the dataset for age

age = [10,12,16,19,11,20,26,28,30,38,35,34,45,60,68,64,62,70,78,75,79,85,94,95]
# creating the histogram
plt.hist(age, bins =[0,20,40,60,80,100])  #bins=5
plt.show()
```

# Pie Chart
Pie Chart is a circular statistical plot that can display only one series of data. The area of the chart is the total percentage of the given data. The area of slices of the pie represents the percentage of the parts of the data. The slices of pie are called wedges. The area of the wedge is determined by the length of the arc of the wedge.

```python
#importing the required libraries
import matplotlib.pyplot as plt
%matplotlib inline

#preparing the dataset
animal = ['Dog','Cat','Wolf','Lion']
count = [50,45,60,80]    #235--->360   # 50/235*100 , if 50/235 the x out of 100

#creating the pie chart
plt.pie(count, labels = animal,autopct='%1.1f%%',shadow=True, startangle = 90,explode = (0,0.2,0,0), colors = ['m','c','b','y']) #25 =25.00
plt.title("Pie Chart Demo")
plt.show()
```

# Donut Chart
Donut charts are the modified version of Pie Charts with the area of center cut out. A donut is more concerned about the use of area of arcs to represent the information in the most effective manner instead of Pie chart which is more focused on comparing the proportion area between the slices. Donut charts are more efficient in terms of space because the blank space inside the donut charts can be used to display some additional information about the donut chart.

For being a Donut chart it must be necessarily a Pie chart. If we look at the pie chart, we will focus on the center of the chart. Donut charts, on the other hand, eliminates the need to compare the size or area of the slice and shifts the focus on the length of the arc, which in turn is easy to measure


```python
import matplotlib.pyplot as plt
%matplotlib inline

#define the dataset
group_names = ["GroupA","GroupB","GroupC"]
group_size = [20,30,50]
size_centre = [10]

a,b,c = [plt.cm.Blues,plt.cm.Reds,plt.cm.Greens]
#Creating the doughnut chart
pie1 = plt.pie(group_size,labels = group_names, autopct='%1.1f%%',radius = 1.5,shadow=True,startangle = 90) #,colors = [a(0.5),b(1.0),c(0.5)])
pie2 = plt.pie(size_centre, radius = 1.0,colors = 'w')
plt.show()
```

# Area Chart
An area plot is like a line graph, but instead of just showing the lines, it fills the space below them with colors. It is a type of data visualization that displays quantitative data using shaded areas to represent different categories or groups over a continuous range.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
#define the dataset
x=range(1,15)  #(1,2,....,14)
y=[1,4,6,8,4,5,3,2,4,1,5,6,8,7]

#creating the area plot
plt.stackplot(x,y,colors = 'g', alpha = 0.4)
#Merge area chart with the line chart
plt.plot(x,y,color = 'k',marker='o')
plt.grid(True)
plt.show()
```

# Box Plot
A box plot, also known as a box-and-whisker plot, provides a visual summary of the distribution of a dataset. It represents key statistical measures such as the median, quartiles and potential outliers in a concise and intuitive manner. Box plots are particularly useful for comparing distributions across different groups or identifying anomalies in the data.

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Data Preparation
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order,discount])
#Plotting the data
plt.boxplot(data, showmeans = True)   #mean value
plt.title("Box Plot Demo")
plt.grid(True)
plt.show()
```
